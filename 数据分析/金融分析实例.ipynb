##### 假如我从2010年1月1日起，每个月的第一个交易日买一手股票，每年的最后一个交易日卖出所有股票，到今天为止，我的收益如何？

##   获取每个月的1号
> resample函数

```python
df_month_first = df2.resample('M').first()  # 按月重新取样，返回每个月的第一天
# 虽然显示的行标签是每个月的最后一天，但是对应的数据是每个月的第一天的数据
# 如果要想行标签显示月份的第一天，则可以将resample的参数设置为 'MS' （month start）
df_month_first
```

```python
df_year_last = df2.resample('A').last()['1998':'2019']
df_year_last
# 由于2020年还没有结束，所以最后显示的2020-12-31实际上是当前时间的数据
#  计算的时候需要提出，2020年没到结束，所以不卖出股票
```

```python
# 开始模拟买股票策略
money = 0   # 个人资产，买股票的钱记录为花出去的钱，卖股票的钱记录为挣的钱
hold = 0    #  持仓股票数
for year in range(1998,2021):
    # 每一年的每个月的首个交易日买1手股票
    money -= df_month_first.loc[str(year),'open'].sum()*100
    hold += len(df_month_first.loc[str(year),'open'])*100
    if year != 2020:
        # 每一年的最后一个交易日卖出股票
        #df_year_last.loc[str(year),'open'] 返回的是Series，不能和股票数相乘，需要获取其浮点数，使用sum，或者索引[0]
        money += hold * df_year_last.loc[str(year),'open'].sum()
        hold = 0
        print(f'{year}收益：{money}')

# 循环结束，还需要计算手上没卖出去的股票的钱
print(f'剩余股票数:{hold}')
money += hold * df_month_first.loc['2020-11','close'].sum()
print(f'总收益为:{money}')
        
```

### 双均线策略

> 导语： 双均线策略，通过建立m天移动平均线，n天移动平均线，则两条均线必有交点。若m>n，n天平均线“上穿越”m天均线则为买入点，反之为卖出点。该策略基于不同天数均线的交叉点，抓住股票的强势和弱势时刻，进行交易。

>对于每一个交易日，都可以计算出前N天的移动平均值，然后把这些移动平均值连起来，成为一条线，就叫做N日移动平均线。
比如前5个交易日的收盘价分别为10,9,9,10,11元，那么，5日的移动平均股价为9.8元。同理，如果下一个交易日的收盘价为12，那么在下一次计算移动平均值的时候，需要计算9,9,10,11,12元的平均值，也就是10.2元。
将这平均值连起来，就是均线。

* 短期均线波动较大，中长期均线波动较小
* 金叉：短期均线上穿长期均线（超过），买入信号
* 死叉：短期均线下穿长期均线，卖出信号

> ### 如果我从2010年1月1日开始，初始资金是10万，金叉尽量买入，死叉全部卖出，则到今天为止，我的炒股收益如何？

```python
df = df['2010':]
df
```

```python
# 金叉
gold_cross = df[(df['ma5'] <= df['ma30']) & (df['ma5'] > df['ma30']).shift(-1)].index
gold_cross[:5]
```

* TTTTTTFFFFFFTTTTTTFFFFF -> ma5 >=ma10
* FFFFFFTTTTTTFFFFFFTTTTT -> ma5 < ma10

>将第二个条件向左移动一格
* TTTTTTFFFFFFTTTTTTFFFFF -> ma5>=ma10
* FFFFFTTTTTTFFFFFFTTTTT -> ma5< ma10

> 可以发现，当两个条件都是True的时候就是死叉出现的时候

```python
# 死叉
dead_cross = df[(df['ma5'] >= df['ma30']) & (df['ma5'] < df['ma30']).shift(-1)].index
dead_cross[:5]
```

```python
money = 100000
gold_ser = pd.Series(1,index=gold_cross)
dead_ser = pd.Series(0,index=dead_cross)
ser = gold_ser.append(dead_ser)
ser.sort_index(inplace=True)
ser # 保存所有的金叉死叉点，并排序好
```

```python
money = 100000  #初始资金
hold = 0   # 持仓股票数

for i in range(len(ser)):
    price = df.loc[ser.index[i],'open']
    if ser.iloc[i] == 0:
        # 死叉卖出
        money += hold*price
        hold = 0
    elif ser.iloc[i] == 1:
        # 金叉买入,需要计算能买多少手
        hands = money // (price*100)
        hold += hands * 100
        money -= price * hold
        
# 如果最后一个是金叉，则还需要计算当前股票的钱
if hold > 0:
    price_now = df.loc[df.index[-1],'open']
    money += hold * price_now
    
print(f'累计收益为:{money - 100000}')      
        
```
