## plot函数：绘制折线图

* ### 线型：linestyle (-,-.,--,:,...)
* ### 点型：marker (v,^,s,*,H,+,x,D,o,...)
* ### 颜色：color (b,g,r,y,k,w,...)

```python
plt.plot([1,2,3,4,5],[2,10,7,8,10],linestyle='--',marker='*',color='r')
plt.show()
```

```python
plt.plot([1,2,3,4,5],[2,10,7,8,10],':Hg') # 三个参数写在一起
plt.show()
```

```python
plt.plot([1,2,3,4,8],[2,10,7,8,10],':Hg') # 三个参数写在一起
plt.plot([2,3,4,7,9],[2,6,7,4,9],'--+r') # 三个参数写在一起
plt.show()
```

### Matplotlib -- 图形标注
* ### 设置图像标题:plt.title()
* ### 设置y轴范围：plt.ylim()
* ### 设置x轴名称：plt.xlabel()
* ### 设置x轴刻度：plt.xticks()
* ### 设置y轴名称：plt.ylabel()
* ### 设置y轴刻度：plt.yticks()
* ### 设置x轴范围: plt.xlim()
* ### 设置曲线图例：plt.legend()

```python
plt.plot([1,2,3,4,5],[2,10,7,8,10],'-Hg',label='lineA') # 设置label属性，用于显示图例
plt.plot([1,2,3,4,5],[4,2,3,5,9],'-*r',label='lineB') 
# 图形标注
plt.title('Test plot')
plt.xlabel('x lable')
plt.ylabel('y lable')
plt.legend()
plt.xticks(np.arange(0,11,2)) # 设置X轴的刻度
plt.show()
```

### 练习，绘制属性函数
> y= x  ;  y = x^2

```python
x = np.linspace(-100,100,1000)
y1 = x.copy()
y2 = x**2

plt.plot(x,y1,color='red',label = 'y = x')
plt.plot(x,y2,color='blue',label = 'y = x^2')

plt.ylim(-100,100)

plt.legend()
plt.show()
```

### 画布与子图 
* 画布:figure    fig = plt.figure()
* 图：subplot  ax1 = fig.add_subplot(2,2,1)
* 调节子图间距： subplots_adjust(left,bottom,right,top,wspace,hspace)

```python
fig = plt.figure()             #创建画布
ax1 = fig.add_subplot(2,2,1)   #创建子图
ax1.plot([1,2,3],[4,5,6])      #添加数据
ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.linspace(-10,10,10),[i for i in range(10)],color='red')
ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.linspace(-10,10,10),[np.random.randint(0,10) for i in range(10)],color='blue')

# fig.show()   在别的编辑器中，调用show()弹出窗口
plt.suptitle('plot name')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(12, 3))

#柱状图
plt.subplot(141)
plt.bar(names, values)
#散点图
plt.subplot(142)
plt.scatter(names, values)
#现状图
plt.subplot(143)
plt.plot(names, values)
#堆叠图
plt.subplot(144)
plt.stackplot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

####  一个plot函数绘制多条线

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0, 5, 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 散点图
>说明：
plt.scatter()
产生一个散点图，前两个参数是x,y轴，c指定颜色，s指定缩放梯度，data指定数据来源。

```python
import numpy as np
import matplotlib.pyplot as plt

data = {'a': np.arange(50),
        'b':np.random.randn(50)*10 + np.random.randint(10),
        'c': np.random.randint(0, 250, 50),
        'd': np.abs(np.random.randn(50))*100
       
       }

plt.scatter(x='a', y='b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

>line, = plt.plot(x, y, '-')
>设置属性：
>line.set_color('r')

>lines = plt.plot(x1, y1, x2, y2)
># use keyword args
>plt.setp(lines, color='r', linewidth=2.0)
