## 读取图片

```python
import matplotlib.pyplot as plt
import numpy as np
# 读取图片
image = cv2.imread(r"Chapter01/01-testing-installation/logo.png") # 彩色图片，返回的是一个三维numpy数组
plt.imshow(image[:,:,::-1],cmap='winter')  #OpenCV颜色模式是BGR，而matplotlib的颜色模式是RGB
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='Greys')  #OpenCV颜色模式是BGR，而matplotlib的颜色模式是RGB
plt.show()
```
